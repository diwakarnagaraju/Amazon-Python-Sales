# Data Cleaning

```python
df.drop(['New', 'PendingS'], axis = 1, inplace = True)
```

```python
df
```

```python
print(df.isna().sum())
```

```python
#converting the datatypes

# Replace NaN values in 'ship-postal-code' before converting to int
df['ship-postal-code'] = df['ship-postal-code'].fillna(0).astype(int)

# Correcting the typo in the date conversion function
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
#Renaming columns for bettere clarity

df.rename(columns = {'Qty' : 'Quantity'}, inplace = True)
```

```python
df.describe(include = 'number')
```

```python
#categorical Data Analysis

for col in df.select_dtypes(include = ['object']):
    print(df[col].unique())
```

Data Visualization


```python
sns.countplot(x = 'Size', data = df,  palette = "coolwarm")
plt.xlabel('size')
plt.ylabel('number of orders')
plt.title('Distribution of Orders by size')
plt.show()
```

```python
#Quantity by Size

size_qty_sum = df.groupby('Size')['Quantity'].sum().reset_index()
sns.barplot(x = 'Size' , y = 'Quantity', data = size_qty_sum, palette = 'plasma')
plt.xlabel('Size')
plt.ylabel('Total Quantity')
plt.title('Total Quantity by size')
plt.show()
```

```python
#Courier Status Distribution

sns.countplot(data = df , x = 'Courier Status' , hue = 'Status', palette = 'magma')
plt.xlabel('Courier Status')
plt.ylabel('Number of Orders')
plt.title('Distribution of Orders by the Courier status')
plt.show()
```

```python
#Category Distribution

df['Category'].value_counts().plot(kind = 'bar', color = sns.color_palette('crest' , len(df['Category'].unique())))
plt.xlabel('Category')
plt.ylabel('Number of Orders')
plt.title('Distribution of orders by categories')
plt.show()                                   
```

```python
#B2B vs Retailer Orders

b2b_counts = df['B2B'].value_counts()
plt.pie(b2b_counts, labels = b2b_counts.index, autopct = '%1.1f%%', colors = sns.color_palette ('Set2'))
plt.title('Distribution of B2B vs Retailer orders')
plt.show()
```

```python
#Fullfilment method

fulfillment_counts = df['Fulfilment'].value_counts()
plt.pie(fulfillment_counts, labels = fulfillment_counts.index, autopct = '%1.1f%%', colors = sns.color_palette('coolwarm'))
plt.title('Distribution of fulfillment methods')
plt.show()
```

```python
#Calculate the purchase frequency by counting the number of oreders per category

df['Purchase Frequency'] = df.groupby('Category')['Order ID'].transform('count')

#Calculate the avg value for each category

df['Average Order Value'] = df.groupby('Category')['Amount'].transform('mean')

#Creating the segmentation based on the avg order value

df['Customer Segment'] = pd.cut(df['Average Order Value'],bins = [-float('inf'), df['Average Order Value'].quantile(0.33), df['Average Order Value'].quantile(0.66), float('inf')], labels = ['Low', 'Medium','High'])
```

```python
df[['Category', 'Purchase Frequency', 'Average Order Value', 'Customer Segment']].head()
```

```python
#scatter plot to show order amount by category for segmentation

plt.figure(figsize = (10,6))
sns.scatterplot(x = 'Category', y = 'Amount', data = df, palette = 'plasma')
plt.xlabel('Category')
plt.ylabel('Amount')
plt.title('Order amount by category')
plt.show()
```

```python
#Sales performance by category

plt.figure(figsize = (10,6))
sns.barplot(x = 'Category', y = 'Amount', data = df, palette = 'coolwarm')
plt.xlabel('Category')
plt.ylabel('Amount')
plt.title('Total sales amount')
plt.xticks(rotation = 45)
plt.show()
```

```python
#sales channel analysis

plt.figure(figsize = (10,6))
sns.barplot(x = 'Sales Channel', y = 'Amount' , data = df, palette = 'crest')
plt.xlabel('Sales Channel')
plt.ylabel('Amount')
plt.title('Total sales by sales channel')
plt.xticks(rotation = 30)
plt.show()
```

```python
#Customer segmentation Anlaysis

plt.figure(figsize = (10,6))
sns.boxplot(x= 'Customer Segment', y = 'Amount', data = df, palette = 'Blues')
plt.xlabel('Customer Segmentation')
plt.ylabel('Amount')
plt.title('Sales distribution by customer segmentation')
plt.show()
```

```python
#Order quantity analysis

plt.figure(figsize = (10,6))
sns.barplot(x = 'Category', y = 'Quantity', data = df, palette = 'viridis')
plt.xlabel('Category')
plt.ylabel('quantity')
plt.title('Total order quantity by categories')
plt.show()
```

```python
# Purchase Frequency vs Avg order Values

plt.figure(figsize = (10,6))
sns.scatterplot(x = 'Purchase Frequency' , y = 'Average Order Value', data = df, hue = 'Customer Segment', palette = 'viridis')
plt.xlabel('Purchase Frequency')
plt.ylabel('Average Order Values')
plt.title('Purchase Frequency vs Avg order Values')
plt.show()
```
